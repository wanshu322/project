```python
# -*- coding: utf-8 -*-
def isEnglish(s):
    """find if a string not English"""
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True
```

```python
def wordFrequency(content):

    """read content(string) and return the dictionary of word frequency
    
     
    """
 

    count_dict ={}



    #remove leading and trailing whitespace
    content = content.strip()
    # remove numbers
    content = re.sub(r'\d+', '', content)
    # remove punctuations and convert characters to lower case
    content = "".join([char.lower() for char in content if char not in string.punctuation])
    content = content.replace("—", "")
    content = content.replace("®", "")
    content = content.replace("·", "")

    #check non-English content
    if isEnglish(content) == True:
        #split the line into words
        words = content.split()



        #increase counters 
        for word in words:
            word = word.lower()
            word = word.translate(str.maketrans('', '', string.punctuation))


            if word not in count_dict and word:
                count_dict[word] = 0
            count_dict[word] +=1
    
  
        



    return count_dict
```

```python
#remove stopwords
def removeStopwords(content):

    stopWords = set(stopwords.words('english'))
    words = word_tokenize(content)
    #wordsFiltered = []
    wordsFiltered = " "
    # remove the stopwords
    for w in words:
        if w not in stopWords:
            #wordsFiltered.append(w)
            wordsFiltered += " " + w

    return wordsFiltered
```

### incorporate basic term weighting - TFxIDF

```python
def transer_TFIDF(df):
    
    df_columns = df.columns
    df_index = df.index
    N = df.shape[1]
    
    df_df  = df.astype(bool).sum(axis=1)
    df_df = pd.DataFrame(df_df, columns =["df"])
    
    # idf = log2(N/nk)
    # df - term in one document frequency 
    df_df["idf"] = np.log2(N/df_df["df"])
    pd.options.display.float_format = '{:,.2f}'.format##IDF table
    idf_arr = df_df["idf"].to_numpy()
    df_arr = df.to_numpy()
    res_arr = np.transpose(df_arr)*idf_arr
    df_tfidf_arr = np.transpose(res_arr)
    pd.DataFrame(df_tfidf_arr,columns = df_columns)
    df_tfidf = pd.DataFrame(df_tfidf_arr,columns = df_columns,index = df_index)
    
    
    
    return df_tfidf
```

```python
def QueryList_toarray(Q,dt_matrix):
    """
    conver the query as a list of terms
    to an array
    input
    Q - a string list with the query words
    dt_matrix - a dataframe with doc-term matrix, the term list as the column names
    
    return an array of boolean """
    
    
    term_list = dt_matrix.columns.tolist()
    Q_arr = np.zeros(len(term_list)) 
    for j in range(len(Q)):
        for i in range(len(term_list)):
            
            if Q[j] ==  term_list[i]:
                Q_arr[i]=1
                   
    
    return Q_arr
```

```python
def CosSim(A,B):
    """compute Cosine similarity for two vectors"""
    num = np.dot(A,B) #dot product
    denom = la.norm(A)*la.norm(B)
    return (num / denom)
```

```python
def DiceSim(A,B):
    """compute Dice Similarity of two vectors (numpy array)"""
    A_index = np.nonzero(A)
    B_index = np.nonzero(B)
    num_index_A = len(A_index[0]) # the number of Ti in A 
    num_index_B = len(B_index[0]) # the number of Ti in B



    sum_index = num_index_A+num_index_B


    index_common = np.intersect1d(A_index[0], B_index[0], assume_unique=False, return_indices=False)
    num_common = len(index_common)



    coef = 2*num_common/sum_index
    
    return coef
```

```python
def JaccardSim(A,B):
    """compute Jaccard Similarity of two vectors (numpy array)"""
    A_index = np.nonzero(A)
    B_index = np.nonzero(B)
    num_index_A = len(A_index[0]) # the number of Ti in A 
    num_index_B = len(B_index[0]) # the number of Ti in B



    sum_index = num_index_A+num_index_B


    index_common = np.intersect1d(A_index[0], B_index[0], assume_unique=False, return_indices=False)
    num_common = len(index_common)
    
    coef = num_common/(sum_index - num_common)

    return coef
```

```python
def query_doc_sim(query_str_lst, doc_term_matrix):
    """compute Jaccard sim (J_sim) and Dice sim (D_sim) and Cosine Sim (Cos_Sim) and return the doc ranking"""
    """input 
    query_str_lst - a list of query words
    doc_term_matrix - a dataframe, which is the original frequency term-document matrix: term-columns, document-rows
    
    output:
    df_doc_sim -  a dataframe with doc name as index, Cosine, Dice, Jaccard similarity as columns 
    ["cos_sim","d_sim","j_sim"]"""
    
    query_arr = QueryList_toarray(query_str_lst,doc_term_matrix) # transfer a query list to an array like

    ##compute similarity 
    num_doc= doc_term_matrix.shape[0]
    
    #initialize the similarity columns
    cos_sim_col = np.zeros(num_doc)
    j_sim_col =np.zeros(num_doc)
    d_sim_col =np.zeros(num_doc)
    
    df_doc_sim = pd.DataFrame(data = np.array([cos_sim_col,j_sim_col,d_sim_col]).T,index = doc_term_matrix.index.tolist(),columns=["cos_sim","d_sim","j_sim"])
    
    
    
    for i in doc_term_matrix.index.tolist():#for each doc, compute the similarity

        df_doc_sim.iloc[i,0]= CosSim(query_arr,doc_term_matrix.iloc[i].to_numpy())#Cosine
        df_doc_sim.iloc[i,1]= DiceSim(query_arr,doc_term_matrix.iloc[i].to_numpy())#Dice
        df_doc_sim.iloc[i,2]= JaccardSim(query_arr,doc_term_matrix.iloc[i].to_numpy())#Jaccard
            
 
    
    return df_doc_sim
```
